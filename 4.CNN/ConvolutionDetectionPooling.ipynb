# Demonstrating Convolution, Detection and Pooling

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
images = []

img = np.zeros((25, 25))
img[:, [12, 13]] = 255
img[[5, 10, 15, 20], :] = -50
images.append(img)

img = np.zeros((25, 25))
img[:, [9, 10]] = 255
img[[5, 10, 15, 20], :] = -50
images.append(img)

img = np.zeros((25, 25))
img[:, [13, 14]] = 255
img[[5, 10, 15, 20], :] = -50
images.append(img)
```

```python
f, subplot = plt.subplots(1, 3, figsize=(15, 15))
for i, img in enumerate(images):
    subplot[i].imshow(img, cmap='gray')
```

# Convolution

```python
# Convolutional Filter

filter = np.array([
    [0, 1, 0],
    [0, 1, 0],
    [0, 1, 0]
])
```

```python
conv_images = []
for img in images:
    convoluted = np.zeros((23, 23))
    for x in range(img.shape[0]-2):
        for y in range(img.shape[1]-2):
            region = img[x:x+3, y:y+3]
            convoluted[x, y] = np.sum(np.multiply(region, filter))/3.0
    conv_images.append(convoluted)
```

```python
convoluted.shape
```

```python
f, subplot = plt.subplots(1, 3, figsize=(15, 15))
for i, img in enumerate(conv_images):
    subplot[i].imshow(img, cmap='gray')
subplot[0].set_title("Hello", size=22)
```

# Activation Function (Detection)

```python
# ReLU Activation Function
relu = lambda x: x * (x > 0)
act_images = []
f, subplot = plt.subplots(1, 3, figsize=(15, 15))

for i, img in enumerate(conv_images):
    img = relu(img)
    subplot[i].imshow(img, cmap='gray')
    subplot[0].set_title("Hello", size=22)
    act_images.append(img)
```

# Pooling 

```python
pooled_images = []
for img in act_images:
    pooled = np.zeros((21, 21))
    for x in range(img.shape[0]-2):
        for y in range(img.shape[1]-2):
            pooled[x, y] = np.max(img[x:x+3, y:y+3])
    pooled_images.append(pooled)
```

```python
f, subplot = plt.subplots(1, 3, figsize=(15, 15))

for i, img in enumerate(pooled_images):
    subplot[i].imshow(img, cmap='gray')
    subplot[0].set_title("Hello", size=22)
```
